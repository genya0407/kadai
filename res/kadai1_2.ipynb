```ruby
require 'daru'
require 'daru_plotting_plotly'

Daru.plotting_library = :plotly
```

```ruby
['0.05','0.15','0.5'].each do |s|
  Daru::DataFrame.from_csv("#{s}.csv").plot(type: :scatter, x: 't', y: ['point 1', 'point 4'], mode: [:markers, :lines]).tap do |plot|
    plot.layout.title = "s=#{s}"
    plot.layout = { font: { size: 22 } }
    plot.layout.yaxis = { title: '存在確率' }
    plot.layout.xaxis = { title: '時刻 t' }
    plot.show
  end
end
nil
```

# 考察

sが大きくなるに連れて、オーバーシュートが大きくなり、また、定常値に落ち着くまでの時間が長くなっている。

sというパラメータは、前の瞬間いた場所に留まる確率を表している。
その場にとどまる確率が高ければ、初めのうちは点１にいる確率が高い。
これによって、点１にいる確率のオーバーシュートが大きくなっている。
また、その場にとどまる確率が高いほど、ある状態から次の状態に移行する確率が低くなり、状態の変化にはより多くの時間がかかると考えられ、これが、定常地に落ち着くまでの時間が高いことに寄与していると考えられる。
