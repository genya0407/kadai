```ruby
require 'daru'
require 'rbplotly'
require 'daru_plotting_plotly'

Daru.plotting_library = :plotly
```

# 課題4.1.1

```ruby
ys = [ 2.672775868652158e-2,
  1.7049097443619377e-4,
  2.8781321762139794e-6,
  5.53866838174516e-8,
  1.1093407078360817e-9,
  2.951160496533767e-11,
  7.038952754001571e-13,
  1.8813076096968473e-14,
  4.77048955893622e-16]
xs = ys.map.with_index { |_, i| i }
Plotly::Plot.new(
  data: [x: xs, y: ys],
  layout: {
    title: '収束判定に用いたスカラ値(片対数)',
    yaxis: { type: :log, title: 'スカラ値' },
    xaxis: { title: 'Aを乗じた回数' }
    }
  ).show
```

# 課題4.2

```ruby
Daru::DataFrame.from_csv('../output/2.csv').tap do |df|
  df.rename_vectors 'trapezoidal' => '複合台形公式', 'simpsons' => '複合シンプソン公式'
end.plot(
          x: 'k',
          y: [
            '複合台形公式',
            '複合シンプソン公式'
          ],
          mode: 'markers+lines',
          layout: {
            title: '複合台形公式と複合シンプソン公式の比較',
            xaxis: { title: 'k' },
            yaxis: { title: '誤差の対数' }
          })
   .show
```
